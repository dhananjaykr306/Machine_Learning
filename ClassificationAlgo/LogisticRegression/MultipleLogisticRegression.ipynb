```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
```

```python
# Load Titanic dataset from seaborn
import seaborn as sns
data = sns.load_dataset('titanic')
```

```python
data.head()
```

```python
# select the feature from data
df = data[['pclass', 'sex', 'age', 'fare','class','survived']]
```

```python
df
```

```python
df.rename(columns={"survived":"target"},inplace=True)
```

```python
df[df.isnull().any(axis=1)]
```

```python
df['age'].fillna(df['age'].mean(),inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df[df.isnull().any(axis=1)]
```

```python
df['class'] = df['class'].map({'Third':3,'Second':2,'First':1})
```

```python
# Convert categorical variable 'sex' into numeric
df['sex'] = df['sex'].map({'male': 0, 'female': 1})
```

```python
# Separate features (X) and target (y)
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Standardize features for better performance
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
# Initialize and train logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
```

```python
# Output results
print(f"Accuracy: {accuracy:.2f}")
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(class_report)
```
