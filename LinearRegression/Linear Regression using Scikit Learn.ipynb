# Data Processing

# spliting the data

```python
# training input and output
train_input = np.array(data.x[0:500]).reshape(500,1)
train_output = np.array(data.y[0:500]).reshape(500,1)

test_input = np.array(data.x[500:699]).reshape(199,1)
test_output = np.array(data.y[500:699]).reshape(199,1)
```

```python
print(f"Training input data shape {train_input.shape}")
print(f"Training output data shape {train_output.shape}")
print(f"Testing input data shape {test_input.shape}")
print(f"Testing input data shape {test_output.shape}")
```

# <font color = 'orange'>Linear Regression</font>

# Training model

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_model = LinearRegression()
```

```python
lr_model = LinearRegression()
lr_model.fit(train_input,train_output)
```

```python
lr_model.coef_
```

```python
lr_model.intercept_
```

# Predictions

```python
test_predictions = lr_model.predict(test_input)
```

```python
test_predictions
```

```python
plt.plot(test_input,test_predictions,'+',color='orange')
plt.plot(test_input,test_output,'*',color='black')
plt.title('Performance Testing')
plt.xlabel("Input")
plt.ylabel("Output Prediction")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
cost = mean_squared_error(test_output,test_predictions)
cost
```
