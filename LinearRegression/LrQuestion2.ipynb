```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
print(train.isnull().sum())
```

```python
train.dropna(inplace=True)
```

```python
print(train.isnull().sum())
```

```python
print(test.isnull().sum())
```

```python
# Initialize the Linear Regression model
model = LinearRegression()
model.fit(train["x"].values.reshape(-1, 1), train["y"].values.reshape(-1, 1))
```

```python
y_pred = model.predict(test['x'].values.reshape(-1,1))
```

```python
# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(test['y'].values.reshape(-1,1), y_pred)
print(f"Mean Squared Error: {mse}")

# Calculate R-squared value
r2 = r2_score(test['y'].values.reshape(-1,1), y_pred)
print(f"R-squared: {r2}")
```

```python
# Scatter plot of actual vs predicted values
plt.figure(figsize=(10,6))
plt.scatter(test['y'].values.reshape(-1,1), y_pred, color='blue')
plt.plot([test['y'].values.reshape(-1,1).min(), test['y'].values.reshape(-1,1).max()], [test['y'].values.reshape(-1,1).min(), test['y'].values.reshape(-1,1).max()], color='red', linestyle='--')
plt.xlabel('Actual Apparent Temperature')
plt.ylabel('Predicted Apparent Temperature')
plt.title('Actual vs Predicted Apparent Temperature')
plt.show()
```

```python
# Get the coefficients for each feature
print("coefficients : ",model.coef_)
```
